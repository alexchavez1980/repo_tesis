```python
import mne                                                              # Librería de python para explorar, visualizar,
mne.set_log_level('WARNING')                                            # y analizar datos neurofisiológicos humanos.
import scipy.io
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb                                                    
color = ['green', 'blue','red','cyan', 'magenta', 'yellow','k','w']     # Paleta de colores para diferenciar las ondas

import a_funciones as a_fun                                             # Funciones Alex
import os                                                               # Para importar varios archivos

xlabel = 'Muestra'                                                      # Abscisas
ylabel = 'Amplitud (uV)'                                                # Ordenadas
```

```python
resultados_amp = pd.read_csv('./a_results/info_final/acc_amp_compilado.csv', index_col=0, delimiter=';')
#resultados_amp
resultados_fase = pd.read_csv('./a_results/info_final/acc_fase_compilado.csv', index_col=0, delimiter=';')
#resultados_fase
resultados_gauss = pd.read_csv('./a_results/info_final/acc_gauss_compilado.csv', index_col=0, delimiter=';')
#resultados_gauss
resultados_gauss_amp = pd.read_csv('./a_results/info_final/acc_gauss_amp_compilado.csv', index_col=0, delimiter=';')
resultados_gauss_amp
```

```python
#--------------------------------------------------------------
#---------- Secuencia de intensificación de amplitud ----------
#--------------------------------------------------------------
#plt.figure(figsize=(30,8))    
plt.figure(figsize=(10,8))                                                                                
axes = plt.gca()                                                    
plt.plot(resultados_amp['Coef_erp'], resultados_amp['acc_P300S01'], 
    linestyle='-.', marker='o', color='red', label='P300S01',
    markerfacecolor='red', markersize=4) 
plt.plot(resultados_amp['Coef_erp'], resultados_amp['acc_P300S02'], 
    linestyle='--', marker='>', color='blue', label='P300S02',
    markerfacecolor='blue', markersize=4) 
plt.plot(resultados_amp['Coef_erp'], resultados_amp['acc_P300S03'], 
    linestyle='-.', marker='o', color='green', label='P300S03',
    markerfacecolor='green', markersize=6) 
plt.plot(resultados_amp['Coef_erp'], resultados_amp['acc_P300S06'], 
    linestyle='--', marker='>', color='orange', label='P300S06',
    markerfacecolor='orange', markersize=6) 
#axes.set_title('Precision')
axes.set_xlabel('Secuencia de intensificación de amplitud (µV)')
axes.set_ylabel('Porcentaje de aciertos')
axes.title.set_size(30)
axes.xaxis.label.set_size(15)
axes.yaxis.label.set_size(15)                                  
plt.grid()
plt.legend(loc='lower right',fontsize='xx-large')
plt.savefig("./a_images/P300S02_accuracy.jpg")
plt.show()
#---------------------------------------------------------
#---------- Secuencia de desplazamiento en fase ----------
#---------------------------------------------------------
#plt.figure(figsize=(30,8))    
plt.figure(figsize=(10,8))                                                                                
axes = plt.gca()                                                    
plt.plot(resultados_fase['Coef_erp'], resultados_fase['acc_P300S01'], 
    linestyle='-.', marker='o', color='red', label='P300S01',
    markerfacecolor='red', markersize=4) 
plt.plot(resultados_fase['Coef_erp'], resultados_fase['acc_P300S02'], 
    linestyle='--', marker='>', color='blue', label='P300S02',
    markerfacecolor='blue', markersize=4) 
plt.plot(resultados_fase['Coef_erp'], resultados_fase['acc_P300S03'], 
    linestyle='-.', marker='o', color='green', label='P300S03',
    markerfacecolor='green', markersize=6) 
plt.plot(resultados_fase['Coef_erp'], resultados_fase['acc_P300S06'], 
    linestyle='--', marker='>', color='orange', label='P300S06',
    markerfacecolor='orange', markersize=6) 

#axes.set_title('Precision')
axes.set_xlabel('Secuencia de desplazamiento en fase (muestras)')
axes.set_ylabel('Porcentaje de aciertos')
axes.title.set_size(30)
axes.xaxis.label.set_size(15)
axes.yaxis.label.set_size(15)                                  
plt.grid()
plt.legend(loc='lower right',fontsize='xx-large')
plt.savefig("./a_images/P300S02_accuracy.jpg")
plt.show()
#-------------------------------------------------------------------------------------
#---------- Inserción de un función gaussiana en reemplazo del template ERP ----------
#-------------------------------------------------------------------------------------
#plt.figure(figsize=(30,8))    
plt.figure(figsize=(10,8))                                                                                
axes = plt.gca()                                                    
plt.plot(resultados_gauss['Coef_erp'], resultados_gauss['acc_P300S01'], 
    linestyle='-.', marker='o', color='red', label='P300S01',
    markerfacecolor='red', markersize=4) 
plt.plot(resultados_gauss['Coef_erp'], resultados_gauss['acc_P300S02'], 
    linestyle='--', marker='>', color='blue', label='P300S02',
    markerfacecolor='blue', markersize=4) 
plt.plot(resultados_gauss['Coef_erp'], resultados_gauss['acc_P300S03'], 
    linestyle='-.', marker='o', color='green', label='P300S03',
    markerfacecolor='green', markersize=6) 
plt.plot(resultados_gauss['Coef_erp'], resultados_gauss['acc_P300S06'], 
    linestyle='--', marker='>', color='orange', label='P300S06',
    markerfacecolor='orange', markersize=6) 
#axes.set_title('Precision')
axes.set_xlabel('Inserción de un función gaussiana en reemplazo del template ERP')
axes.set_ylabel('Porcentaje de aciertos')
axes.title.set_size(30)
axes.xaxis.label.set_size(15)
axes.yaxis.label.set_size(15)                                  
plt.grid()
plt.legend(loc='lower right',fontsize='xx-large')
plt.savefig("./a_images/P300S02_accuracy.jpg")
plt.show()
#-------------------------------------------------------------------------------------
#---------- Inserción de un función gaussiana en reemplazo del template ERP ----------
#---------------------- Secuencia de intensificación de amplitud ---------------------
#-------------------------------------------------------------------------------------
#plt.figure(figsize=(30,8))    
plt.figure(figsize=(10,8))                                                                                
axes = plt.gca()                                                    
plt.plot(resultados_gauss_amp['Coef_erp'], resultados_gauss_amp['acc_P300S01'], 
    linestyle='-.', marker='o', color='red', label='P300S01',
    markerfacecolor='red', markersize=4) 
plt.plot(resultados_gauss_amp['Coef_erp'], resultados_gauss_amp['acc_P300S02'], 
    linestyle='--', marker='>', color='blue', label='P300S02',
    markerfacecolor='blue', markersize=4) 
plt.plot(resultados_gauss_amp['Coef_erp'], resultados_gauss_amp['acc_P300S03'], 
    linestyle='-.', marker='o', color='green', label='P300S03',
    markerfacecolor='green', markersize=6) 
plt.plot(resultados_gauss_amp['Coef_erp'], resultados_gauss_amp['acc_P300S06'], 
    linestyle='--', marker='>', color='orange', label='P300S06',
    markerfacecolor='orange', markersize=6) 
#axes.set_title('Precision')
axes.set_xlabel('Sec. de intensificación de amplitud con función gaussiana en reemplazo del template ERP')
axes.set_ylabel('Porcentaje de aciertos')
axes.title.set_size(30)
axes.xaxis.label.set_size(15)
axes.yaxis.label.set_size(15)                                  
plt.grid()
plt.legend(loc='lower right',fontsize='xx-large')
plt.savefig("./a_images/P300S02_accuracy.jpg")
plt.show()
```
