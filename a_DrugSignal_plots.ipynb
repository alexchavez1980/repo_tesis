# Gráfica completa del df_eeg

```python
files_path_itba_test = ['./dataset/itba/P300S01.mat', './dataset/itba/P300S02.mat']
xlim_general=([7000,15500])
ylim_general=([-50,80])
for i, path in enumerate(files_path_itba_test):   
    df_eeg = pd.read_csv(f'./a_results/df_eeg_P300S0{i+1}.csv')
    eje_x_sample=df_eeg['sample']

    fig, (ax0, ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8, ax9) = plt.subplots(nrows=10, ncols=1,figsize=(30,20))
    fig.suptitle(f"DrugSignal con t_stim y t_type  \n Eje x: muestra. Eje y: uV \n Paciente P300S0{i+1}", fontsize=30)

    ax0.plot(eje_x_sample, df_eeg.iloc[:,0], label='0') 
    ax0.set_xlim(xlim_general), ax0.set_ylim(ylim_general), ax0.set_title(df_eeg.columns[0]), ax0.grid()

    ax1.plot(eje_x_sample, df_eeg.iloc[:,1], label='1')
    ax1.set_xlim(xlim_general), ax1.set_ylim(ylim_general), ax1.set_title(df_eeg.columns[1]), ax1.grid()

    ax2.plot(eje_x_sample, df_eeg.iloc[:,2], label='2')
    ax2.set_xlim(xlim_general), ax2.set_ylim(ylim_general), ax2.set_title(df_eeg.columns[2]), ax2.grid()

    ax3.plot(eje_x_sample, df_eeg.iloc[:,3], label='3')
    ax3.set_xlim(xlim_general), ax3.set_ylim(ylim_general), ax3.set_title(df_eeg.columns[3]), ax3.grid()

    ax4.plot(eje_x_sample, df_eeg.iloc[:,4], label='4')
    ax4.set_xlim(xlim_general), ax4.set_ylim(ylim_general), ax4.set_title(df_eeg.columns[4]), ax4.grid()

    ax5.plot(eje_x_sample, df_eeg.iloc[:,5], label='5')
    ax5.set_xlim(xlim_general), ax5.set_ylim(ylim_general), ax4.set_title(df_eeg.columns[5]), ax5.grid()

    ax6.plot(eje_x_sample, df_eeg.iloc[:,6], label='6')
    ax6.set_xlim(xlim_general), ax6.set_ylim(ylim_general), ax6.set_title(df_eeg.columns[6]), ax6.grid()

    ax7.plot(eje_x_sample, df_eeg.iloc[:,7], label='7')
    ax7.set_xlim(xlim_general), ax7.set_ylim(ylim_general), ax7.set_title(df_eeg.columns[7]), ax7.grid()

    ax8.plot(eje_x_sample, df_eeg.iloc[:,8], label='8')
    ax8.set_xlim(xlim_general), ax8.set_ylim([-3,14]), ax8.set_title(df_eeg.columns[8]), ax8.grid()

    ax9.plot(eje_x_sample, df_eeg.iloc[:,9], label='9')
    ax9.set_xlim(xlim_general), ax9.set_ylim([-0.2,2.2]), ax9.set_title(df_eeg.columns[9]), ax9.grid()
    
    plt.savefig(f"./a_images/df_eeg - Paciente P300S0{i+1}.jpg")
    #plt.show()
```
