#### Análisis de la señal trials  
*data.trial: Sample point where each of the 35 trials starts*: https://www.kaggle.com/datasets/rramele/p300samplingdataset  
No hay nada mas que un vector.  
Se obtiene de p300-subject-25.mat, igual que signal.
¿Por qué de p300-subject-25.mat y no de P300S04.mat por ejemplo?: Lo veo abajo.    
El ERPTemplate.mat es para obetner los ERPs (Event-related potential) y drogar la señal.  

```python

mat = scipy.io.loadmat('./dataset/p300-subject-25.mat')                         # Now load the basal EEG stream 
                                                                                # t_trials = mat['data'][0][0][3] -> El profesor lo trae como array.
t_trials = pd.DataFrame(mat['data'][0][0][3])                                   # Lo paso directamente a dataframe.
t_trials
t_trials = t_trials.transpose()                                                 # Lo vuelvo columna.
t_trials['sample'] = t_trials.index                                             # Le agrego la columna sample del index para graficarlo.
plt.figure(figsize=(30,8))
plt.plot(t_trials['sample'], t_trials[0])
plt.grid(), plt.show()
t_trials.head()
t_trials.describe()   
```
