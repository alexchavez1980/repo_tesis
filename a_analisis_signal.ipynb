#### Análisis de la señal Signal (p300-subject-25.mat)
Signal proviene de p300-subject-25.mat  :  p300subject25['data'][0][0][0].  
Se obtiene de p300-subject-25.mat, igual que t_trials.  
Tiene 8 canales.


```python
p300subject25 = scipy.io.loadmat('./dataset/p300-subject-25.mat')
signal = a_fun.to_df(p300subject25['data'][0][0][0])
signal.head(5)
```

```python
# Gráfico de uno de los canales
plt.figure(figsize=(30,8))
plt.plot(signal['sample'], signal[0], color[1])
plt.xlim([0,70000])
plt.ylim([-200,250])
plt.grid(), plt.show()
```

```python
# Gráfico de todos los canales.
plt.figure(figsize=(30,8))
for i in range(0,6):
    plt.plot(signal['sample'], signal[i], color[i])
plt.grid()
plt.savefig("07 - P300-subject-25.jpg")
plt.show()
```

```python
# Gráfico de todos los canales en una ventana específica
plt.figure(figsize=(30,8))
axes = plt.gca()
for i in range(0,6):
    plt.plot(signal['sample'], signal[i], color[i])
plt.ylim([-7000, 5000])
plt.xlim([225000,240000])
plt.grid()
plt.savefig("08 - p300-subject-25-WINDOW01.jpg")
plt.show()

# Gráfico de todos los canales en una ventana específica
plt.figure(figsize=(30,8))
axes = plt.gca()
for i in range(0,6):
    plt.plot(signal['sample'], signal[i], color[i])
plt.ylim([-500, 500]);
plt.xlim([236000,238000])
plt.grid()
plt.savefig("04 - p300-subject-25-WINDOW02.jpg")
plt.show()
```
